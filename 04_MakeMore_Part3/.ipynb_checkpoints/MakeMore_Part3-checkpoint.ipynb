# MakeMore Part 3: Activations & Gradients, BatchNorm
* To understand more complex architectures, we need to go back to MLPs and build an intuitive understanding of activations and gradients
* Especially how gradients behave and look like
* Recurring Neural Networks are very expressive in that they are an universal approximator and can in principle implement all the algorithms
* But RNNs are not very easily optimizable with the first-order gradient based techniques that we have available to us
* The key to understanding, why they are not easily optimizable is the behavior of the activations and the gradients and how they behave during training
* We will see that a lot of the variants since RNNs have tried to improve on that situation

## The (cleaned-up) MLP Code

````python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
words = open('names.txt', 'r').read().splitlines()
words[:8]
````

````python
len(words)
````

````python
# build the vocabulary of characters and mappings to/from integers
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
vocab_size = len(itos)
print(stoi)
print(vocab_size)
````

````python
# build the dataset
block_size = 3 # context length: how many characters do we take to predict the next one?

print(len(words))
def build_dataset(words):
    X, Y = [], []
    
    for w in words:
        context = [0] * block_size
        for ch in w + '.':
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:]+[ix]
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape)
    return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])
````

* The parameter initialization is cleaned-up a bit, as the magic numbers are removed and turned into variables

````python
# MLP revisited
def initParams():
    global g, C, W1, b1, W2, b2, parameters
    n_embd = 10 # the dimensionality of the character embedding vectors
    n_hidden = 200 # the number of neurons in the hidden layer of the MLP
    
    g = torch.Generator().manual_seed(2147483647) # for reproducibility
    C = torch.randn((vocab_size, n_embd),           generator=g)
    W1 = torch.randn((n_embd*block_size, n_hidden), generator=g)
    b1 = torch.randn(n_hidden,                      generator=g)
    W2 = torch.randn((n_hidden, vocab_size),        generator=g)
    b2 = torch.randn(vocab_size,                    generator=g)
    
    parameters = [C, W1, b1, W2, b2]
    print(sum(p.nelement() for p in parameters)) # number of parameters in total
    for p in parameters:
        p.requires_grad = True
````

* Training of the net

````python

def train():
    # same optimization as last time
    global C, W1, b1, W2, b2
    max_steps = 200000
    batch_size = 32
    global lossi
    lossi = []
    global logits
    global hpreac
    
    for i in range(max_steps):
        
        # minibatch construct
        xi = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
        Xb, Yb = Xtr[xi], Ytr[xi]                                              # batch X,Y
    
        # forward pass
        emb = C[Xb]                                                            # embed the charaters into vectors
        embcat = emb.view(emb.shape[0], -1)                                    # concatenate the vectors
        hpreac = embcat @ W1 + b1                                              # hidden layer pre-activation
        h = torch.tanh(hpreac)                                                 # hidden layer
        logits = h @ W2 + b2                                                   # output layer
        loss = F.cross_entropy(logits, Yb)                                     # loss function
        
        # backward pass
        for p in parameters:
            p.grad = None
        loss.backward()
    
        # update
        lr = 0.1 if i < 100000 else 0.01                                       # step learning rate decay
        for p in parameters:
            p.data += -lr * p.grad
    
        # track stats
        if i % 10000 == 0:                                                     # print every once in a while
            print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')

        lossi.append(loss.log10().item())    
````

````python
initParams()
train()
````

* Plot the loss of batch per iteration

````python
plt.plot(lossi)
````

* Print training and validation loss for the full splits

````python
@torch.no_grad()
def split_loss(split):
    x, y = {
        'train': (Xtr, Ytr),
        'val': (Xdev, Ydev),
        'test': (Xte, Yte),
    }[split]
    emb = C[x]
    embcat = emb.view(emb.shape[0], -1) # concat into (N, block_size * n_embd)
    h = torch.tanh(embcat @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, y)
    print(split, loss.item())

split_loss('train')
split_loss('val')
````

* Sample from the trained model

## Better Initialization of the Parameters
* The first line printed while training was ```0 / 200000: 27.8817```
* It comes rapdidly down to $2$, so we can tell that the initial loss is way too high
* In training of NN it is almost always the case that we will have a rough idea of what loss to expect at initialization
* That depends on the loss function and the problem setup
* In this case we expect a much lower number and we can calculate that
* In our problem, there are 27 characters that could come next for any one training example
* There is no reason to believe any character would come more likely than the others
* So we would expect that the probability distribution that comes out initially is the uniform distribution
* This means that any character should have a probability of roughly $1/27$
* The loss is the average negative log likelihood, which takes the negative log of each probability, adds them up and divides by the number of samples
* With all probabilities equal, it will be the same as the loss of just one number:

````python
-torch.tensor(1/27).log()
````

* So we would expect a loss of $3.29$
* Drawing the initial probabilities from the random distribution will give high probability to the wrong characters

### Example for 4 Characters

* Let's imagine we only have 4 characters with a very wrong distribution, i.e. high numbers assigned to the wrong character and a low number to the right one

````python
logits = torch.tensor([0.0, 0.0, 0.0, 0.0])
probs = torch.softmax(logits, dim=0)
loss = -probs[2].log()
probs, loss
````

* When all predictions are equal, i.e. form a diffuse distribution as above, the probabilities will all be equal and we'd expect a loss of $1.3$
* As the logits will start to change, the loss will change, too
* If by chance the correct character received a high logit value, the loss will be low

````python
logits = torch.tensor([0.0, 0.0, 5.0, 0.0])
probs = torch.softmax(logits, dim=0)
loss = -probs[2].log()
probs, loss
````

* But is is more likely that some other logit will receive a high value and then we will record a much higher loss

````python
logits = torch.tensor([0.0, 5.0, 0.0, 0.0])
probs = torch.softmax(logits, dim=0)
loss = -probs[2].log()
probs, loss
````

* If the logits go to more extreme values, it is more unlikely to pick the correct one and the loss will become even higher

````python
logits = torch.randn(4) * 100
probs = torch.softmax(logits, dim=0)
loss = -probs[2].log()
probs, loss
````

* So we want the logits to be roughly equal when the network is initialized

````python
logits = torch.tensor([1.0, 1.0, 1.0, 1.0])
probs = torch.softmax(logits, dim=0)
loss = -probs[2].log()
probs, loss
````

````python
def trainOnce():
    # same optimization as last time
    max_steps = 200000
    batch_size = 32
    global h
    global lossi
    lossi = []
    global logits
    global hpreac
    
    for i in range(max_steps):
        # minibatch construct
        xi = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
        Xb, Yb = Xtr[xi], Ytr[xi]                                              # batch X,Y
    
        # forward pass
        emb = C[Xb]                                                            # embed the charaters into vectors
        embcat = emb.view(emb.shape[0], -1)                                    # concatenate the vectors
        hpreac = embcat @ W1 + b1                                              # hidden layer pre-activation
        h = torch.tanh(hpreac)                                                 # hidden layer
        logits = h @ W2 + b2                                            # output layer
        loss = F.cross_entropy(logits, Yb)                                     # loss function
        
        # backward pass
        for p in parameters:
            p.grad = None
        loss.backward()
    
        # update
        lr = 0.1 if i < 100000 else 0.01                                       # step learning rate decay
        for p in parameters:
            p.data += -lr * p.grad
    
        # track stats
        if i % 10000 == 0:                                                     # print every once in a while
            print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')
        lossi.append(loss.log10().item())
        
        break    
        
````

### Changing the Parameter Initialization
* All parameters are initialized randomly
* We start at the end and see that we shouldn't add anything at the end, i.e. set ```b2``` to zero
* Also we will scale down the ```W2``` weights by a factor of $0.01$

````python
# MLP revisited
def initParamsZero():
    global g, C, W1, b1, W2, b2, parameters
    n_embd = 10
    n_hidden = 200
    
    g = torch.Generator().manual_seed(2147483647)
    C = torch.randn((vocab_size, n_embd),           generator=g)
    W1 = torch.randn((n_embd*block_size, n_hidden), generator=g)
    b1 = torch.randn(n_hidden,                      generator=g)
    W2 = torch.randn((n_hidden, vocab_size),        generator=g) * 0.01         # Scale down W1 weights
    b2 = torch.randn(vocab_size,                    generator=g) * 0            # Initialize b2 to all zeros
    
    parameters = [C, W1, b1, W2, b2]
    print(sum(p.nelement() for p in parameters))
    for p in parameters:
        p.requires_grad = True
````

* We'll explain in a second why we cannot set the ```W2``` weights to exactly zero

````python
initParamsZero()
trainOnce()
````

* The loss is now close enough, but not exactly what we would get from a uniform distribution
* This is because our weights have some entropy, which will be used for symmetry breaking as we'll see in a minute
* The logits will now come out much closer to zero

## The Problem with the hidden Layer Output
* The problem lies within the activations of the hidden state ```h``` in our training code

* We will understand the problem with this activation, once we understand how gradients flow through the NN

### The Problem during Backpropagation
* The hidden layer is computed as
  
    ```h = torch.tanh(hpreact)```

* The layer is made up of 200 neurons for each one of the samples and the layer implements an element-wise tanh
* Let's look at what happens with tanh in the backward pass
* When we implemented micrograd in the first lecture, the backpropagation for tanh looked as follows

  ```
    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x)+1) 
        out = Value(t, (self, ), 'tanh')
        def backward():
            self.grad += (1-t**2)*out.grad
        out._backward = backward
        return out
  ```
* The backward pass computes the gradient with the formula $(1-t^2)*out.grad$, where $t$ is $tanh$ itself
* if $tanh$ is $-1.0$ or $1.0$, then the whole expression becomes zero no matter what $out.grad$ is
* The gradient will basically be destroyed at a $tanh$ node with activations farther than 1 from 0
* Looking at an activation of exactly zero, the term $(1-t^2)$ will become exactly $1.0$ and the gradient is just passed through
* So going from zero into the positive or negative direction will only ever decrease the gradient until it vanishes

### The dead Neuron

* Remember that h contains the activations for all samples (the rows) and how those samples activate each of the 200 neurons (the columns)
* Each sample activates the neurons differently and might cause the neuron's activation to land in the flat part of tanh, yielding -1 or 1 as the output
* In this case, the weights of the neuron will not be changed during gradient descent
* If for a specific neuron all samples will make the neuron's output to land in the flat part, then that neuron will never learn and is called a *dead neuron*
* Let's inspect the hidden layer output ```h``` for dead neurons

````python
h.shape
````

````python
plt.figure(figsize=(20,10))
plt.imshow(h.abs() > 0.99, cmap='gray', interpolation='nearest')
````

* We don't see a completely white column in the last iteration and last batch
* So every neuron will learn something from the batch
* The dead neuron effect might also happen with other non-linearities

* Sigmoid is also a squashing neuron and will have the same problem
* ReLU will have a derivative of 0 below 0 and 1 above zero ($f(x)=x$, so $\frac{df}{dx}=1$, for $x > 0$)
* So if a ReLU neuron never activates for a given training set, it will never learn
* Leaky ReLU doesn't have a flat tale, so it will not suffer from this issue as much
* ELU again has flat parts, so it might suffer from this issue

* In summary, dead neurons might happen at initialization, which we can avoid (shown in the next section)
* But dead neurons can even happen during training
* If you have a too high learning rate for example, some neurons might get to high of a gradient and get knocked out off the data manifold

## Scale Weight Matrix to avoid dead Neurons
* Coming back to the preactivations after initialization, we can see the the values are spread out too extreme

````python
plt.hist(hpreac.view(-1).tolist(), 50);
````

* All values outside of $(-1,1)$ will be squashed to $-1$ and $1$ respectively and will land in the flat zone of $tanh$
* For all those values the network will not update the neuron
* We want most of the preactivations to land in the right range, so backprop will have an effect on that neuron
* The way we can achieve this is to scale the weights matrix and the bias of the hidden layer

````python
# MLP revisited
def initParamsZeroH():
    global g, C, W1, b1, W2, b2, parameters
    n_embd = 10
    n_hidden = 200
    
    g = torch.Generator().manual_seed(2147483647)
    C = torch.randn((vocab_size, n_embd),           generator=g)
    W1 = torch.randn((n_embd*block_size, n_hidden), generator=g) * 0.2
    b1 = torch.randn(n_hidden,                      generator=g) * 0.01
    W2 = torch.randn((n_hidden, vocab_size),        generator=g) * 0.01         # Scale down W1 weights
    b2 = torch.randn(vocab_size,                    generator=g) * 0            # Initialize b2 to all zeros
    
    parameters = [C, W1, b1, W2, b2]
    print(sum(p.nelement() for p in parameters))
    for p in parameters:
        p.requires_grad = True
````

* Now we check how the outputs of the hidden layer are distributed

````python
initParamsZeroH()
trainOnce()
````

* And that is because of the preactivations, which are kept at unit gaussian

## Choosing Scaling at Initialization more formally
* We chose a randomly small scaling factor to bring both weights matrices closer to zero
* But we can't rely on magic numbers and need to find a more formal way to choose these
* Let's look at an example, where we choose a random training set with 1000 samples, each of dimension 10
* We will also initialize a random weights matrix with matching dimensionality and 200 neurons

````python
x = torch.randn(1000, 10)
w = torch.randn(10, 200)
y = x @ w
print(x.mean(), x.std())
print(y.mean(), y.std())
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.hist(x.view(-1).tolist(), 50, density=True);
plt.subplot(122)
plt.hist(y.view(-1).tolist(), 50, density=True);
````

* multiplying by $w$ increases the standard deviation of the result $y$, but we want a standard deviation of $1$ throughout the net, so gradients are not blocked
* As we saw, we can scale $w$ to achieve this, but how to choose the correct scaling factor?
* When you work out of the variance of the multiplication, you will need to divide $w$ by the square root of the fan-in

````python
x = torch.randn(1000, 10)
w = torch.randn(10, 200)  / 10**0.5  # dividing by sqrt of fan-in
y = x @ w
print(x.mean(), x.std())
print(y.mean(), y.std())
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.hist(x.view(-1).tolist(), 50, density=True);
plt.subplot(122)
plt.hist(y.view(-1).tolist(), 50, density=True);
````

* A number of papers have looked into how to best initialize MLPs
* MLPs can have fairly deep networks that can have these non-linearities in between
* We want to make sure these activations are well-behaved and they don't expand to infinity or shrink to zero
* The question is how we initialize the weights, so these non-linearities take on reasonable values throughout the network
* One of these papers is [Delving Deep into Recitfiers](https://arxiv.org/abs/1502.01852), where they study the ReLU and PReLU non-linearity instead of $tanh$, but the analysis is very similar
* 
* ReLU clamps everything that is negative to zero
* Because you basically throw away half of the gaussian distribution, you need to compensate for that with a gain
* We multiply by $\sqrt{1/10}$, while they have to scale by the factor of $2$ as they discard half of the distibution: $\sqrt{2/10}$
* But they not only study the behavior of the activations in the forward pass of the network, but also that the gradients are well behaved, because we will update our weights
* Basically with this initialization, the backward pass will also work well
* pytorch has $torch.nn.init.kaiming\_normal()$, which normalizes either the activations or the gradients, which doesn't make too much of a difference, so we can leave the default
* You also have to specify the nonlinearity
* Depending on the type of non-linearity, the input will be multiplied with a different gain in addition to the normalization
* This gain will intuitively fight the squashing of the non-linearities, so if we have many layers that apply e.g. a tanh, then the distribution will be squashed a bit more
* The gain will compensate for this and was chosen empirically by pytorch to be $5/3$

### Scale at Initialization not that important

* We skimmed through the initialization problem, because modern innovations reduce the dependency on good initialization
* These modern innovations are normalization layers, like batch normalization and better optimizers like ADAM
* In practice it is enough to normalize the initial weights with $5/3 * \sqrt{fan\_in}$

````python
# Kaiming Normalization
def initParamsKaiming():
    global g, C, W1, b1, W2, b2, parameters
    n_embd = 10
    n_hidden = 200
    stddev_normalization = (5/3)/((n_embd * block_size)**0.5)
    
    g = torch.Generator().manual_seed(2147483647)
    C = torch.randn((vocab_size, n_embd),           generator=g)
    W1 = torch.randn((n_embd*block_size, n_hidden), generator=g) * stddev_normalization
    b1 = torch.randn(n_hidden,                      generator=g) * 0.01
    W2 = torch.randn((n_hidden, vocab_size),        generator=g) * 0.01         # Scale down W1 weights
    b2 = torch.randn(vocab_size,                    generator=g) * 0            # Initialize b2 to all zeros
    
    parameters = [C, W1, b1, W2, b2]
    print(sum(p.nelement() for p in parameters))
    for p in parameters:
        p.requires_grad = True
````

## Batch Normalization
* One of the methods that made the network more stable wrt the initalization was batch normalization introduced in the google paper [Batch Normalization: Accelerating Deep Network Training by Reducing Internal Covariate Shift](https://arxiv.org/abs/1502.03167)
* It made it possible to train very deep neural nets very reliably, so we are going to explain it here
* Within our training code, we don't want the hidden states to get too close to zero, because then the layer wouldn't activate at all
* But we also don't want it to become too saturated, as saturated neurons don't learn anymore
* We want all of them to be roughly gaussian (zero mean with a unit std deviation)
* At least we want this at initialization
* The insight from the paper is that we can add normalization layers to each layer and make normalization part of the training
* So normalization happens after the linear layer ($hpreac$) and before the $tanh$ layer ($h$)
  
  ```
  hpreac = embcat @ W1 + b1
  # normalize here
  h = torch.tanh(hpreac)
  ```
* To normalize, we need the current mean and std deviation of $hpreac$

````python
hpreac.shape
````

````python
hpreac.mean(0, keepdim=True).shape
````

````python
hpreac.std(0, keepdim=True).shape
````

* To use these values, we look at the prescription in the paper:

* Basically calculate the mean and std dev of all outputs for one neuron
* Subtract mean and divide by std dev for all outputs

````python

def trainBN():
    # same optimization as last time
    global C, W1, b1, W2, b2
    max_steps = 200000
    batch_size = 32
    global lossi
    lossi = []
    global logits
    global hpreac
    
    for i in range(max_steps):
        
        # minibatch construct
        xi = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
        Xb, Yb = Xtr[xi], Ytr[xi]                                              # batch X,Y
    
        # forward pass
        emb = C[Xb]                                                            # embed the charaters into vectors
        embcat = emb.view(emb.shape[0], -1)                                    # concatenate the vectors
        hpreac = embcat @ W1 + b1                                              # hidden layer pre-activation
        hpreac = (hpreac - hpreac.mean(0, keepdim=True)) / (hpreac.std(0, keepdim=True))
        h = torch.tanh(hpreac)                                                 # hidden layer
        logits = h @ W2 + b2                                                   # output layer
        loss = F.cross_entropy(logits, Yb)                                     # loss function
        
        # backward pass
        for p in parameters:
            p.grad = None
        loss.backward()
    
        # update
        lr = 0.1 if i < 100000 else 0.01                                       # step learning rate decay
        for p in parameters:
            p.data += -lr * p.grad
    
        # track stats
        if i % 10000 == 0:                                                     # print every once in a while
            print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')

        lossi.append(loss.log10().item())    
````

* This will however not give us a very good result, because we want the preactivations to be roughtly gaussian, but only at initialization
* We don't want them to be forced to be gaussian throughout training
* We would like the NN to move this around to make it more diffuse more sharp, make some neurons more trigger happy or less trigger happy
* To enable this, the paper also introduces two parameters $\gamma$ and $\beta$ to learn the scaling and shifting of the distributions 

````python
# Kaiming Normalization
def initParamsKaimingBN():
    global g, C, W1, b1, W2, b2, bngain, bnbias, parameters
    n_embd = 10
    n_hidden = 200
    stddev_normalization = (5/3)/((n_embd * block_size)**0.5)
    
    g = torch.Generator().manual_seed(2147483647)
    C = torch.randn((vocab_size, n_embd),           generator=g)
    W1 = torch.randn((n_embd*block_size, n_hidden), generator=g) * stddev_normalization
    b1 = torch.randn(n_hidden,                      generator=g) * 0.01
    W2 = torch.randn((n_hidden, vocab_size),        generator=g) * 0.01         # Scale down W1 weights
    b2 = torch.randn(vocab_size,                    generator=g) * 0            # Initialize b2 to all zeros

    bngain = torch.ones((1, n_hidden))
    bnbias = torch.zeros((1, n_hidden))
    
    parameters = [C, W1, b1, W2, b2, bngain, bnbias]
    print(sum(p.nelement() for p in parameters))
    for p in parameters:
        p.requires_grad = True
````

````python

def trainBN():
    # same optimization as last time
    global C, W1, b1, W2, b2
    max_steps = 200000
    batch_size = 32
    global lossi
    lossi = []
    global logits
    global hpreac
    
    for i in range(max_steps):
        
        # minibatch construct
        xi = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
        Xb, Yb = Xtr[xi], Ytr[xi]                                              # batch X,Y
    
        # forward pass
        emb = C[Xb]                                                            # embed the charaters into vectors
        embcat = emb.view(emb.shape[0], -1)                                    # concatenate the vectors
        hpreac = embcat @ W1 + b1                                              # hidden layer pre-activation
        hpreac = bngain * (hpreac - hpreac.mean(0, keepdim=True)) / hpreac.std(0, keepdim=True) + bnbias
        h = torch.tanh(hpreac)                                                 # hidden layer
        logits = h @ W2 + b2                                                   # output layer
        loss = F.cross_entropy(logits, Yb)                                     # loss function
        
        # backward pass
        for p in parameters:
            p.grad = None
        loss.backward()
    
        # update
        lr = 0.1 if i < 100000 else 0.01                                       # step learning rate decay
        for p in parameters:
            p.data += -lr * p.grad
    
        # track stats
        if i % 10000 == 0:                                                     # print every once in a while
            print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')

        lossi.append(loss.log10().item())    
````

* Let's train and look at the loss

````python
initParamsKaimingBN()
trainBN()
````

* Before we calculate the loss, we need to add the bn parameters to the split loss function

````python
@torch.no_grad()
def split_loss_bn(split):
    x, y = {
        'train': (Xtr, Ytr),
        'val': (Xdev, Ydev),
        'test': (Xte, Yte),
    }[split]
    emb = C[x]
    embcat = emb.view(emb.shape[0], -1) # concat into (N, block_size * n_embd)
    hpreac = embcat @ W1 + b1
    hpreac = bngain * (hpreac - hpreac.mean(0, keepdim=True)) / hpreac.std(0, keepdim=True) + bnbias
    h = torch.tanh(hpreac)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, y)
    print(split, loss.item())
````

### Batch Normalization introduced Batch Dependencies
* Before we introduced batch normalization, each output of the neural net was dependent on one sample alone
* Now that we introduced BN, the mean and std deviation of the whole batch is used to modify the results after the linear layer
* On top of this, the samples in the batch are selected randomly, so depending on which samples come for a ride in the batch, the activation will look slightly differently
* So if we look at one sample, the outputs of the activation layer and the output layer will jitter slightly, depending on the rest of the batch selected at random
* This might seem like a bug or something undesirable, but it turns out that this jitter has the effect of a regularizer
* The introduced jitter will basically pad out any one of the input samples and it is introducing a little bit of entropy at the output
* Because of the padding-out it is a form of data-augmentation, which we will get to later
* The samples are augmented, which makes it harder for a network to overfit to the samples
* On the other hand nobody likes this property, because it usually introduces bugs
* This is the reason there have been many attempts to find other methods of normalization like instance normalization or group normalization

### Batch Normalization and Inference
* Once we finished training, we'd like to feed in a single sample and get a prediction back
* But how do we do this, if the NN in the forward pass estimates the statistics of the mean and the std dev of the batch
* The paper has the following proposal for this: after training we will have an additional step that sets the mean and std dev of the training set a single time
* The follwing code calculates this

````python
with torch.no_grad():
    # pass the training set through
    emb = C[Xtr]
    embcat = emb.view(emb.shape[0], -1)
    hpreac = embcat @ W1 + b1
    # measure the mean/std over the entire training set
    bnmean = hpreac.mean(0, keepdim=True)
    bnstd = hpreac.std(0, keepdim=True)
````

* Now we can adjust split_loss to use these precomputed values

````python
@torch.no_grad()
def split_loss_bn(split):
    x,y = {
        'train': (Xtr, Ytr),
        'val': (Xdev, Ydev),
        'test': (Xte, Yte),
    }[split]
    emb = C[x]
    embcat = emb.view(emb.shape[0], -1)
    hpreac = embcat @ W1 + b1
    hpreac = bngain * (hpreac - bnmean) / bnstd + bnbias # using precalculated values
    h = torch.tanh(hpreac)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, y)
    print(split, loss.item())

split_loss_bn('train')
split_loss_bn('val')
````

* We get the same result, but we can now forward a single example

### Running Mean and Std Dev
* To avoid a postprocessing step, we can also calculate the mean and std dev on the whole training set in a running manner
* The paper suggests to calculate this, by keeping a running mean and std dev and adding the mean and std dev of the current iteration to the running variables with a scaling factor
* We need to initialize them to unit gaussian, because it needs to reflect what W1 is at initialization

````python
# Running Mean / Std Dev
def initParamsRunning():
    global g, C, W1, b1, W2, b2, bngain, bnbias, parameters
    global bnmean_running, bnstd_running                      # introduce the running vars
    
    n_embd = 10
    n_hidden = 200
    stddev_normalization = (5/3)/((n_embd * block_size)**0.5)
    
    g = torch.Generator().manual_seed(2147483647)
    C = torch.randn((vocab_size, n_embd),           generator=g)
    W1 = torch.randn((n_embd*block_size, n_hidden), generator=g) * stddev_normalization
    b1 = torch.randn(n_hidden,                      generator=g) * 0.01
    W2 = torch.randn((n_hidden, vocab_size),        generator=g) * 0.01
    b2 = torch.randn(vocab_size,                    generator=g) * 0

    bngain = torch.ones((1, n_hidden))
    bnbias = torch.zeros((1, n_hidden))

    bnmean_running = torch.zeros((1, n_hidden))     # init running mean to zeros
    bnstd_running = torch.ones((1, n_hidden))     # init running std dev to ones
    
    parameters = [C, W1, b1, W2, b2, bngain, bnbias]
    print(sum(p.nelement() for p in parameters))
    for p in parameters:
        p.requires_grad = True
````

* In the training, we will update it with each iteration, but it will be kept outside of the gradient descent, so we can decorate it with $no\_grad$

````python

def trainRunning():
    # same optimization as last time
    global C, W1, b1, W2, b2, bngain, bnbias
    global bnmean_running, bnstd_running
    max_steps = 200000
    batch_size = 32
    global lossi
    lossi = []
    global logits
    global hpreac
    
    for i in range(max_steps):
        
        # minibatch construct
        xi = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
        Xb, Yb = Xtr[xi], Ytr[xi]
    
        # forward pass
        emb = C[Xb]
        embcat = emb.view(emb.shape[0], -1)
        hpreac = embcat @ W1 + b1
        bnmeani = hpreac.mean(0, keepdim=True)                      # storing iteration's mean in a variable
        bnstdi = hpreac.std(0, keepdim=True)                        # storing iteration's std dev in a variable
        hpreac = bngain * (hpreac - bnmeani) / bnstdi + bnbias

        with torch.no_grad():
            bnmean_running = 0.999 * bnmean_running + 0.001 * bnmeani
            bnstd_running = 0.999 * bnstd_running + 0.001 * bnstdi
        
        h = torch.tanh(hpreac)
        logits = h @ W2 + b2
        loss = F.cross_entropy(logits, Yb)
        
        # backward pass
        for p in parameters:
            p.grad = None
        loss.backward()
    
        # update
        lr = 0.1 if i < 100000 else 0.01
        for p in parameters:
            p.data += -lr * p.grad
    
        # track stats
        if i % 10000 == 0:
            print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')

        lossi.append(loss.log10().item())    
````

* We will now initialize and train, while keeping track of the running mean and std deviation
* The result should match what we calculated in the postprocessing step, only now it is integrated in the training

````python
initParamsRunning()
trainRunning()
````

* Let's compare

````python
torch.cdist(bnmean_running, bnmean)
````

````python
torch.cdist(bnstd_running, bnstd)
````

* Similar enough
* Now we can use these values in the split loss function, instead of the ones from the postprocessing step

````python
@torch.no_grad()
def split_loss_bn_running(split):
    x,y = {
        'train': (Xtr, Ytr),
        'val': (Xdev, Ydev),
        'test': (Xte, Yte),
    }[split]
    emb = C[x]
    embcat = emb.view(emb.shape[0], -1)
    hpreac = embcat @ W1 + b1
    hpreac = bngain * (hpreac - bnmean_running) / bnstd_running + bnbias # using precalculated values
    h = torch.tanh(hpreac)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, y)
    print(split, loss.item())

split_loss_bn('train')
split_loss_bn('val')
````

* Last notes:
* There is an $\epsilon$ in the normalization formula that is added to the variance
* That $\epsilon$ prevents a division by zero, if the variance happens to be zero
* Also, we add a bias to the hidden layer, but we center the distribution to zero afterwards
* Thus, we the bias does nothing and we can get rid of it
* The bnbias now takes care of bias learning

# Summary of Batch Normalization
* We are using batch normalization to control the statistics of activations in the NN
* It is common to sprinkle BN layers across the net
* And usually we will place it after layers that have multiplications, like a linear layer or a conv layer (will cover in the future)
* The BN internally has parameters for the gain and the bias, which are trained using backprop
* It also has two buffers, which are the running mean and std dev
* These are not trained using backprop but are trained using this janky/running update
* The update is done by calculating the statistics of the activations that is feeding into the BN layer over the batch
* Then it is centering that batch to be unit gaussian and then it is offseting and scaling it by the learned parameters
* On top of that it is keeping track of the mean and std dev of the inputs and its maintaining the running mean and the std dev to be used at inference

## Helpful Histograms
* It is helpful to look at the histograms of:
  - activations: shouldn't be too saturated and should not shrink by tanh
  - gradients: should also be roughly unit gaussian and should not expand through layers
  - gradient to data ratio: gradients that are added to the weights should be roughly a 1000 times smaller than the data itself
* The histogram of the last layer will look different, though, and will update with gradients which are only a tenth of the magnitude of the weights
* The reason is that we scaled the weights at initialization to make it less confident
* This will stabilize during training
* When the gradient to data ratio is way below 1e-3, then the network is learning too slow and might indicate a too small learning rate
